```python
import cv2
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import shutil
```

```python
# DIRECTORIES
path_data = "data/"
path_chess_data = path_data + "chess/"
path_real = path_data + "real_pictures/"
path_good = path_real + "Good/"
path_bad = path_real + "Bad/"
path_cad = path_data + "CAD_pictures/"
path_train = path_chess_data + "train/"
path_val = path_chess_data + "val/"

# Create Directories
shutil.rmtree(path_chess_data) # Delete previous chess dataset directory with its contents
os.makedirs(path_chess_data) 
os.makedirs(path_train)
os.makedirs(path_val)

# PARAMETERS
img_size = 256
```

```python
for filename in os.listdir(path_cad):
    if filename.endswith('.png'):
        name = filename.replace('.png', '')
        # load cad picture and resize it to square
        cad_img = cv2.resize(cv2.imread(path_cad + filename, cv2.IMREAD_COLOR), (img_size , img_size))
        for img_name in os.listdir(path_good): 
            if name == img_name[:-7]:
                # load and resize real image
                real_img = cv2.resize(cv2.imread(path_good + img_name, cv2.IMREAD_COLOR), (img_size , img_size))
                # concatenate real and cad images in the same picture
                final_img = cv2.hconcat([real_img, cad_img])
                # save the result to file with the right name and folder structure
                if int(img_name[-6:-4]) == 1:
                    path = path_val + name + img_name[-7:-4] + '.jpg'
                else: 
                    path = path_train + name + img_name[-7:-4] + '.jpg'
                cv2.imwrite(path, final_img)
                print(path)
```
